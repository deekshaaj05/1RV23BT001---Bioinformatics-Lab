```python
import os
from google.colab import files
!pip install biopython
from Bio import SeqIO, Entrez
import requests
import io
from IPython.display import display
from Bio.SeqRecord import SeqRecord
```

```python
mysearch = Entrez.esearch(db='nucleotide',term="[cytochrome oxidase]")
myrecord = Entrez.read(mysearch)
mysearch.close()
myids=myrecord["IdList"]
for id1 in myids:
  Entrez.email='ajdeeksha.bt23@rvce.edu.in'
  filename = id1 + ".fasta"
  if not os.path.isfile(filename):
    net_handle = Entrez.efetch( db='nucleotide', id=id1, rettype="fasta", retmode="text" )
    out_handle= open(filename, "w")
    out_handle.write(net_handle.read())
    files.download(filename)
    out_handle.close()
    net_handle.close()
  myrecord = SeqIO.read(filename,"fasta")
  print(myrecord)
```

```python
for filename in os.listdir():
    if filename.endswith(".fasta") and "_mRNA" not in filename and "_RNA" not in filename:
      record = SeqIO.read(filename, "fasta")
      mrna_seq = record.seq.transcribe()
      mrna_record = SeqRecord(
            mrna_seq,
            id=record.id,
            name=record.name,
            description=record.description
        )
      print(f"ID: {mrna_record.id}")
      print(f"Name: {mrna_record.name}")
      print(f"Description: {mrna_record.description}")
      print(f"Number of features: {len(mrna_record.features)}")
      print(mrna_record.seq)
      print("")

      mrna_filename = filename.replace(".fasta", "_mRNA.fasta")
      SeqIO.write(mrna_record, mrna_filename, "fasta")

      files.download(mrna_filename)
```
