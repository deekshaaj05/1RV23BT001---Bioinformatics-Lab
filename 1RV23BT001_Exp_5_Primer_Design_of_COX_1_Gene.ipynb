```python
from Bio import Entrez, SeqIO
import primer3

Entrez.email = "your.email@example.com"
handle = Entrez.efetch(db="nucleotide", id="PX448095.1", rettype="fasta", retmode="text")
record = SeqIO.read(handle, "fasta")
handle.close()
sequence = str(record.seq)

primers = primer3.bindings.design_primers(
    {'SEQUENCE_TEMPLATE': sequence},
    {
        'PRIMER_OPT_SIZE': 20,
        'PRIMER_MIN_SIZE': 18,
        'PRIMER_MAX_SIZE': 25,
        'PRIMER_OPT_TM': 60,
        'PRIMER_MIN_TM': 57,
        'PRIMER_MAX_POLY_X': 4,
        'PRIMER_MAX_TM': 63
    }
)

forward_primer = primers['PRIMER_LEFT_0_SEQUENCE']
reverse_primer = primers['PRIMER_RIGHT_0_SEQUENCE']

print("Forward Primer:", forward_primer)
print("Reverse Primer:", reverse_primer)
```

```python
from Bio.SeqUtils import gc_fraction, MeltingTemp

#For actual seuqneces from Primer-Blast, paste sequences after uncommenting
#forward_primer = 'sequence'
#reverse_primer = 'sequence'

# 1. GC Content (%)
gc_forward = gc_fraction(forward_primer) * 100
gc_reverse = gc_fraction(reverse_primer) * 100

print(f"Forward Primer GC Content: {gc_forward:.2f}%\n")
print(f"Reverse Primer GC Content: {gc_reverse:.2f}%\n")

# 2. Melting Temperature (Tm) using Wallace rule
tm_forward = MeltingTemp.Tm_Wallace(forward_primer)
tm_reverse = MeltingTemp.Tm_Wallace(reverse_primer)

print(f"Forward Primer Tm: {tm_forward:.2f}°C\n")
print(f"Reverse Primer Tm: {tm_reverse:.2f}°C\n")

# Evaluation
print("Primer Evaluation ")

# GC content check
if 40 <= gc_forward <= 60 and 40 <= gc_reverse <= 60:
    print("GC content for both primers is within the ideal range (40–60%).")
else:
    print("GC content for one or both primers is outside the ideal range (40–60%).")

# Tm compatibility check
if abs(tm_forward - tm_reverse) <= 5:
    print("Tm difference between forward and reverse primers is acceptable (≤ 5°C).")
else:
    print("Tm difference between forward and reverse primers is too large (> 5°C). Consider redesigning.")
```
