```python
from Bio import Entrez, SeqIO

Entrez.email = "your.email@example.com"
handle = Entrez.efetch(db="nucleotide", id="PX448096.1", rettype="fasta", retmode="text")
record = SeqIO.read(handle, "fasta")
handle.close()
sequence = str(record.seq)

# Define restriction enzymes and recognition sequences
enzymes = {
    "HindIII": "AAGCTT",
    "EcoRI": "GAATTC",
    "KpnI": "GGTACC"
}

# Search for restriction sites
restriction_sites = {}

for enzyme, site in enzymes.items():
    positions = []
    index = sequence.find(site)
    while index != -1:
        positions.append(index + 1)  # +1 for biological indexing
        index = sequence.find(site, index + 1)
    restriction_sites[enzyme] = positions

# Display results
for enzyme, positions in restriction_sites.items():
    print(f"{enzyme} restriction sites:")
    if positions:
        print("Positions:", positions)
    else:
        print("No sites found")
    print()
```
